# Question 3 
Implement function triangleArea(a,b,c) that takes as input the lengths of the 3  
sides of a triangle and returns the area of the triangle. By Heron's formula, the area  
of a triangle with side lengths a, b, and c is s(s − a)(s −b)(s −c), where s = (a + b + c)/2.

```
using System;

public class Triangle
{
    double Tri_area(float x,float y,float z)
    {
        double s = (x+y+z)/2;
        double area_of_triangle  = Math.Sqrt(s * (s - x) *(s - y) * (s - z));
        return area_of_triangle;
    }
    public static void Main(string[] args)
    {
        Triangle T = new Triangle();
        Console.WriteLine("Enter the first side's length:");
        double a = Convert.ToDouble(Console.ReadLine());
        Console.WriteLine("Enter the second side's length:");
        double b = Convert.ToDouble(Console.ReadLine());
        Console.WriteLine("Enter the third side's length:");
        double c = Convert.ToDouble(Console.ReadLine());
        Console.WriteLine(T.Tri_area(2,2,2));
    }
}
```

# Question 4
Write a program in C# Sharp to separate odd and even integers in separate arrays.  
Go to the editor  
Test Data :  
Input the number of elements to be stored in the array :5  
Input 5 elements in the array :  
element - 0 : 25  
element - 1 : 47  
element - 2 : 42  
element - 3 : 56  
element - 4 : 32 

```
using System;

public class Seperation
{
    public static void Main(string[] args)
    {
        Console.Write("Input the number of elements to be stored in the array:");
        int Number_of_elements = Convert.ToInt32(Console.ReadLine());
        Console.WriteLine("Input {0} elements in the array :",Number_of_elements);
        
        int[] Array_of_elements = new int[Number_of_elements];
        int even=0;
        int odd=0;
        
        int []Array_even = new int[Number_of_elements];
        int []Array_odd = new int[Number_of_elements];
        
        for (int i = 0;i<Number_of_elements;i++)
        {
            Console.Write("\n element {0}-:",i);
            
            Array_of_elements[i] = Convert.ToInt32(Console.ReadLine());
            
            if(Array_of_elements[i]%2==0)
            {
                Array_even[even] = Array_of_elements[i]; 
                even++;
            }
            else
            {
                Array_odd[odd] = Array_of_elements[i];
                odd++;
            }
        }
        Console.WriteLine("The Even integers are:");
        for (int i = 0;i<even;i++)
        {
            Console.Write(Array_even[i]+"\n");
        }
        Console.WriteLine("The Odd integers are: ");
        for (int i = 0;i<odd;i++)
        {
            Console.Write(Array_odd[i]+"\n");
        }
    }
}
```
